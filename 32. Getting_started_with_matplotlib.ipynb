```python
# Create a Simple Plot using Pyplot

import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 5, 8, 12])
Y = np.array([3, 6, 7, 12, 9])

plt.plot(X,Y)
plt.show()
```

```python
# Creating Bar Graphs using Pyplot

subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
marks = [97, 68, 59, 81, 77, 92]

# We can create horizontal bar graphs using the barh method instead of the bar method.
plt.bar(subjects, marks, color='blue')

plt.title("Bar Graph Example")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()
```

```python
# Creating Clustered Bar Graphs using Pyplot

subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
student1 = [97, 68, 59, 81, 77, 92]
student2 = [88, 61, 80, 40, 62, 52]
student3 = [54, 62, 77, 54, 71, 98]

index = np.arange(6)
width = 0.03


plt.bar(index, student1, width, color="aqua", label="Student 1")
plt.bar(index + width, student2, width, color="green", label="Student 2")
plt.bar(index + (width*2), student3, width, color="blue", label="Student 3")

plt.title("Stacked Bar Graph Example")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(index + width/2, subjects)
plt.legend()
plt.show()

```

```python
# Creating Pie Charts using Pyplot

firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_shares = [10, 40, 30, 5, 15]

Explode = [0, 0, 0, 0.5, 0]

plt.pie(market_shares, explode=Explode, labels=firms, shadow=True, startangle=45)
plt.axis('equal')
plt.legend(title = "List of Firms")

plt.show()

```

```python
# Creating Histograms using Pyplot

x = np.random.randn(10000)

plt.title("Histogram Example")
plt.xlabel("Random Data")
plt.ylabel("Frequency")

plt.hist(x,10)
plt.show()
```

```python
# Create 3D Projections using Matplotlib

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)


figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
# Create a Scatter Plot using Pyplot

import matplotlib.pyplot as plt
from numpy.random import rand

for color in ['red', 'green', 'blue']:
    n = 100
    x, y = rand(2, n)
    scale = 500.0 * rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.09)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.62, 0.62, 0.3, 0.3])

y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine') 
axes2.set_title("cosine") 
plt.show()
```
